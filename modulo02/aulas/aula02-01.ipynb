#Compreensão de Listas X expressões geradoras


A primeira diferença está na sintaxe

 - Compreensões de Listas: São definidas usando colchetes [].




 - Expressões Geradoras: São definidas usando parênteses ()

Também temos a diferença no tipo de resultado

 - Compreensões de Listas: Elas criam listas, ou seja, os valores resultantes são armazenados em memória como uma lista completa.

 - Expressões Geradoras: Elas criam objetos geradores, que são sequências preguiçosas de valores e não armazenam todos os valores em memória de uma vez. Os valores são calculados sob demanda.


 O gerador produz um item por vez e gera o item apenas quando em demanda. Considerando que, em uma compreensão de lista, Python reserva memória para a lista inteira. Assim, podemos dizer que as expressões geradoras são mais eficientes em termos de memória do que as listas.



```python
from sys import getsizeof #Essa função clacula o tamanho do objeto em bytes

comp = [i for i in range(10000)]
gen = (i for i in range(10000))
x = getsizeof(comp)
print("x = ", x)
y = getsizeof(gen)
print("y = ", y)
```

No código acima, podemos ver que os tamanhos dos objetos têm diferença considerável.

```python
list_comprehension = [i for i in range(11) if i % 2 == 0]

print(list_comprehension)
```

```python
generator_expression = (i for i in range(11) if i % 2 == 0)

print(generator_expression)
```

Um objeto gerador (ou simplesmente gerador) é uma estrutura de dados em Python que permite a geração sob demanda de valores a partir de uma função. Os geradores são usados para criar sequências de valores de maneira eficiente, economizando memória, especialmente quando se lida com grandes conjuntos de dados ou fluxos contínuos de informações.

No exemplo acima, se quisermos imprimir a saída das expressões geradoras, podemos simplesmente iterar sobre o objeto gerador.

```python
for i in generator_expression:
    print(i, end=" ")
```

## Yield

Dentro de uma função o yield funciona mais ou menos como um return, com a diferença que ele retorna um generator. O yield guarda o ultimo estado da função geradora para continuar do mesmo lugar novamente.

1 - Quando a função geradora é chamada, ela não começa a ser executada imediatamente. Em vez disso, ela retorna um objeto gerador que é uma espécie de "promessa" de produzir valores quando necessário.

2 - Quando você itera sobre o gerador (por exemplo, usando um loop for), a função é executada até encontrar a instrução yield.

3 -Quando o yield é alcançado, o valor após o yield é produzido como saída do gerador. O estado da função é "pausado" temporariamente.

4 - Na próxima iteração (quando você chama next() ou continua o loop), a função continua a partir do ponto onde parou, executando o código após o yield e produzindo o próximo valor.

```python
def gerador_de_numeros():
    yield 1
    yield 2
    yield 3

gen = gerador_de_numeros()
```

```python
print(next(gen))  # Produz e imprime 1
```
